# Sistem Persamaan Linier

## pengertian sistem persamaan linier
Sistem Persamaan Linier (SPL) adalah kumpulan dari dua atau lebih persamaan linier yang memiliki variabel yang sama dan harus diselesaikan secara bersamaan.

contoh persamaan 3 variabel:

$x + y + z = 3$
$2x - y + z = 4$
$x + 2y - z = 1$

bentuk umum SPL:

$a₁x + b₁y + c₁z = d$₁
$a₂x + b₂y + c₂z = d₂$
$a₃x + b₃y + c₃z = d₃$

## Jenis Penyelesaian SPL ada  3:
1. Tunggal: Hanya satu solusi (potong di satu titik)
2. Tak Hingga: Banyak solusi (grafik berupa garis/himpunan yang berimpit)
3. Tidak Ada Solusi: Tidak ada titik temu (garis sejajar/tidak konsisten)

## metode penyelesaian SPL
a. metode substitusi

cara penyelesaaian menggunakan metode substitusi:
Pilih salah satu persamaan, kemudian nyatakan salah satu variabel dalam bentuk variabel lainnya.
Substitusikan hasil tersebut ke persamaan lain.
Selesaikan persamaan untuk menemukan nilai variabel yang tersisa.

b. metode eliminasi

cara penyelesaaian menggunakan metode eliminasi:
Samakan salah satu koefisien variabel dengan mengalikan persamaan jika diperlukan.
Kurangkan atau jumlahkan kedua persamaan untuk menghilangkan salah satu variabel.
Selesaikan persamaan yang tersisa.

c.eliminasi gauss

Eliminasi Gauss bertujuan mengubah matriks augmented menjadi bentuk eselon baris atas (upper triangular form) menggunakan operasi baris elementer.
cara penyelesaaian menggunakan metode eliminasi:
Bentuk segitiga atas (eliminasi ke bawah),
Ubah elemen-elemen di bawah diagonal utama menjadi nol menggunakan operasi baris.
Substitusi mundur (back-substitution),
Selesaikan nilai variabel dari persamaan yang sudah disederhanakan.




## penjelasan Tugas 

### jika berhimpitam mempunyai banyak solusi

persamaan:
$x+2y+3z=6
2x+4y+6z=12
3x+6y+9z=18.$    

Langkah 1: Analisis Koefisien
Perhatikan bahwa setiap persamaan merupakan kelipatan dari persamaan pertama:

Baris kedua adalah 2 kali baris pertama.
Baris ketiga adalah 3 kali baris pertama.
Artinya, ketiga persamaan tersebut sebenarnya merepresentasikan persamaan yang sama. Oleh karena itu, sistem persamaan ini tidak memiliki solusi unik tetapi memiliki banyak solusi (tak hingga solusi).

Langkah 2: Representasi dalam Matriks
Sistem persamaan di atas dapat ditulis dalam bentuk matriks augmented:

$\begin{bmatrix} 
1 & 2 & 3 & \vert & 6 \\ 
2 & 4 & 6 & \vert & 12 \\ 
3 & 6 & 9 & \vert & 18 
\end{bmatrix}$

Kita lakukan eliminasi baris:

Baris kedua dikurangkan 2 kali baris pertama:

$R_2 \leftarrow R_2 - 2R_1$

Hasilnya:

$\begin{bmatrix} 
1 & 0 & 3 & \vert & 6 \\ 
2 & 0 & 6 & \vert & 0 \\ 
3 & 0 & 9 & \vert & 18 
\end{bmatrix}$

Baris ketiga dikurangkan 3 kali baris pertama:

$R_3 \leftarrow R_3 - 3R_1$

Hasilnya:

$\begin{bmatrix} 
1 & 0 & 0 & \vert & 6 \\ 
2 & 0 & 0 & \vert & 0 \\ 
3 & 0 & 0 & \vert & 0 
\end{bmatrix}$
Langkah 3: Menentukan Solusi
Dari matriks terakhir, kita hanya memiliki satu persamaan independen:

$x+2y+3z=6$
Karena ada tiga variabel tetapi hanya satu persamaan, maka kita bisa menyatakan solusi dalam bentuk parametrik:

Biarkan y = t dan z =s sebagai parameter bebas.
Dari persamaan pertama:
$x=6−2y−3z$

Substitusi y=t dan z=s:
$x=6−2t−3s$


Kesimpulan
Solusi dari sistem persamaan ini dalam bentuk parametrik adalah:

$\begin{aligned}
x &= 6 - 2t - 3s \\
y &= t \\
z &= s
\end{aligned}$

dengan t,s sebagai bilangan real bebas.
Karena ada dua parameter bebas (t dan s),ini menunjukkan bahwa sistem memiliki tak hingga banyak solusi yang membentuk bidang dalam ruang tiga dimensi.

### jika tidak berpotongan tidak ada solusi

persamaan:
$x+y+z=1
x+y+z=3
x+y+z=5$

Langkah 1: Perhatikan Struktur Persamaan
Semua persamaan memiliki bentuk yang sama, yaitu $x+y+z$, tetapi dengan hasil yang berbeda (1, 3, dan 5). Ini tidak mungkin terjadi karena satu ekspresi tidak dapat memiliki tiga nilai berbeda pada saat yang sama.

Langkah 2: Kurangkan Persamaan
Misalnya, kita kurangkan persamaan kedua dengan persamaan pertama:
$\begin{align*}
(x + y + z) - (x + y + z) &= 3 - 1 \\
0 &= 2
\end{align*}$

Ini merupakan kontradiksi (pernyataan yang salah).

Lakukan hal yang sama untuk persamaan ketiga dengan persamaan pertama:
$\begin{align*}
(x + y + z) - (x + y + z) &= 5 - 1 \\
0 &= 4
\end{align*}$

Ini juga kontradiksi.

Kesimpulan
Sistem ini adalah inkonsisten, artinya tidak ada solusi.
Dalam konteks geometri, ini menunjukkan bahwa tiga bidang yang direpresentasikan oleh persamaan tersebut sejajar tetapi tidak berpotongan, sehingga tidak ada titik perpotongan yang memenuhi semua persamaan sekaligus.

### jika berpotongan mempunyai satu penyelesaian tunggal.

persamaanya:
$x+y+z=6
2x-y+3z=14
3x+2y-z=2$

kita bisa menggunakan metode eliminasi atau substitusi. Berikut langkah-langkah penyelesaiannya:

Langkah 1: Eliminasi Salah Satu Variabel
Kita akan menghilangkan salah satu variabel, misalnya y.
Eliminas y dari Persamaan (1) dan (3)
Menjumlahkan (1) dan (3):


$(x + y + z) + (3x + 2y - z) = 6 + 2$

$4x + 3y = 8 \quad \text{(Persamaan A)}$
Eliminasi y dari Persamaan (2) dan (3)
Mengalikan Persamaan (1) dengan 2 agar y sejajar dengan Persamaan (2):
$2(x + y + z) = 2(6)$

$2x + 2y + 2z = 12$
Lalu kita kurangkan dari Persamaan (2):

$(2x - y + 3z) - (2x + 2y + 2z) = 14 - 12$
$-3y + z = 2 \quad \text{(Persamaan B)}$

Langkah 2: Substitusi dan Eliminasi Lanjutan
Kita punya sistem:


$4x + 3y = 8$

$-3y + z = 2$

Dari Persamaan B:

$z = 3y + 2$

Substitusikan \( z \) ke Persamaan (1):

$x + y + (3y + 2) = 6$

$x + 4y = 4 \quad \text{(Persamaan C)}$

Substitusikan \( z = 3y + 2 \) ke Persamaan (3):

$3x + 2y - (3y + 2) = 2$
$3x + 2y - 3y - 2 = 2$

$3x - y = 4 \quad \text{(Persamaan D)}$

Langkah 3: Menyelesaikan Sistem Dua Variabel

Kita punya:

$x + 4y = 4$

$3x - y = 4$

Eliminasi \( x \) dengan mengalikan Persamaan C dengan 3:

$3(x + 4y) = 3(4)$

$3x + 12y = 12$
Kurangkan dari Persamaan D:

$(3x + 12y) - (3x - y) = 12 - 4$

$13y = 8$


$y = \frac{8}{13}$

Substitusikan \( y \) ke Persamaan C:

$x + 4 \left(\frac{8}{13}\right) = 4$

$x + \frac{32}{13} = 4$
$x = 4 - \frac{32}{13}$

$x = \frac{52}{13} - \frac{32}{13}$

$x = \frac{20}{13}$

Substitusikan $( y = \frac{8}{13} \)$ ke Persamaan B:

$z = 3\left(\frac{8}{13}\right) + 2$
$z = \frac{24}{13} + 2$

$z = \frac{24}{13} + \frac{26}{13}$

$z = \frac{50}{13}$

$\textbf{Jawaban Akhir}$ 

$(x, y, z) = \left( \frac{20}{13}, \frac{8}{13}, \frac{50}{13} \right)$

Jadi, solusi dari sistem persamaan ini adalah:

$x = \frac{20}{13}, \quad y = \frac{8}{13}, \quad z = \frac{50}{13}$

# Transformasi Matriks 

## Transformasi Matriks 

Transformasi matriks adalah fungsi yang mengubah vektor dari satu ruang vektor ke ruang vektor lain menggunakan perkalian matriks.

Misalnya, jika 𝐴 adalah matriks 𝑚 × 𝑛 dan $\vec{x} ∈ 𝑅^𝑛$ maka transformasi 𝑇 didefinisikan sebagai:

$T(\vec{x}) = A\vec{x}$

Outputnya adalah vektor $\vec{y}∈ 𝑅^𝑚.$
Transformasi seperti ini disebut transformasi linear, dan merupakan bagian penting dalam pemetaan antara ruang vektor.

### a.  Perkalian Matriks-Vektor
Fokus pada vektor 2D $(𝑅^2)$ untuk mempermudah visualisasi.
Matriks 2 × 2 bisa digunakan untuk mengubah vektor 2D dan tetap menghasilkan vektor 2D.

Contoh matriks:
$A = \begin{bmatrix} 1 & 4 \\ 2 & 3 \end{bmatrix}$

jika $\vec{x}$ = $\begin{bmatrix} 1 \\ 0 \end{bmatrix}$ maka:

$A\vec{x} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$

Visualisasi digunakan untuk menunjukkan bagaimana vektor sebelum dan sesudah transformasi berubah pada bidang kartesius.

### b. Transformasi Dasar
Beberapa contoh transformasi umum yang bisa direpresentasikan oleh matriks:

Refleksi terhadap sumbu-x: $\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$

Refleksi terhadap sumbu-y: $\begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}$

Rotasi berlawanan arah jarum jam sebesar 𝜃 : $\begin{bmatrix} \cos \theta & -\sin \theta \\ \sin \theta & \cos \theta \end{bmatrix}$

Peregangan horizontal atau vertikal: 

Horizontal: $\begin{bmatrix}
k & 0 \\
0 & 1
\end{bmatrix}$ , 
Vertikal: $\begin{bmatrix}
1 & 0 \\
0 & k
\end{bmatrix}$

### c. Komposisi Transformasi
Transformasi dapat dikombinasikan dengan mengalikan matriks secara berurutan.
Misal, transformasi T1 diikuti T2:
$A = A_2 \cdot A_1$

Maka transformasi total $T(\vec{x}) = A_2 (A_1 \vec{x})$

### d. Latihan (Exercises)
Latihan diberikan berupa sketsa transformasi dan identifikasi matriks yang melakukannya. Contoh tugas:

Temukan matriks yang mengubah bujur sangkar satuan dengan rotasi atau refleksi tertentu.
Contoh transformasi majemuk:

1. Refleksi horizontal diikuti refleksi vertikal.

2. Peregangan horizontal lalu refleksi terhadap garis 𝑦=𝑥

# tugas 1 membuktikan $T(v_1, v_2) = (v_1 + v_2, v_1)$ adalah transformasi linier 

Untuk membuktikan bahwa transformasi $T(v_1, v_2) = (v_1 + v_2, v_1)$ adalah transformasi linier, kita harus menunjukkan bahwa:

1. Penjumlahan vektor dipertahankan: $T(u+v)=T(u)+T(v)$ 
2. Perkalian skalar dipertahankan: $T(c⋅v)=c⋅T(v)$

Misalkan: $u = (u_1, u_2),\quad v = (v_1, v_2),\quad c \in \mathbb{R}$

1.  Cek Penjumlahan: 

$T(u + v) = T((u_1 + v_1, u_2 + v_2)) = ((u_1 + v_1) + (u_2 + v_2),\ u_1 + v_1)$

$= (u_1 + u_2 + v_1 + v_2,\ u_1 + v_1)$

Sementara: 

$T(u) = (u_1 + u_2,\ u_1),\quad T(v) = (v_1 + v_2,\ v_1)$

$T(u) + T(v) = (u_1 + u_2 + v_1 + v_2,\ u_1 + v_1)$

Sama, jadi penjumlahan dipertahankan.

2. Cek Perkalian Skalar:

$T(c \cdot v) = T((cv_1, cv_2)) = (cv_1 + cv_2,\ cv_1) = c(v_1 + v_2,\ v_1)$

$= c \cdot T(v)$

Sama, jadi perkalian skalar juga dipertahankan.

Karena kedua sifat dipenuhi, maka:

$T(v_1, v_2) = (v_1 + v_2,\ v_1)$
adalah transformasi linier.

# Refleksi

**Refleksi matriks** (reflection matrix) adalah **matriks transformasi** yang mencerminkan (memantulkan) titik-titik dalam ruang terhadap **sumbu, garis, bidang, atau titik tertentu**.

Refleksi mengubah arah satu atau lebih komponen vektor, seperti bayangan di cermin.

---

### ✅ **Repleksi pada 2D (dua dimensi)**:

1. **Refleksi terhadap sumbu-x**:

$$
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
x \\
-y
\end{bmatrix}
$$

→ Membalik titik terhadap sumbu-x (nilai y dibalik, x tetap).

2. **Refleksi terhadap sumbu-y**:

$$
\begin{bmatrix}
-1 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
-x \\
y
\end{bmatrix}
$$

→ Membalik titik terhadap sumbu-y.

3. **Refleksi terhadap garis y = x**:

$$
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
y \\
x
\end{bmatrix}
$$

→ Menukar posisi x dan y.

---

### ✅ ** Repleksi pada 3D (tiga dimensi)**:

Refleksi bisa dilakukan terhadap bidang seperti $x=0$, $y=0$, atau $z=0$. Contoh refleksi terhadap bidang $x=0$:

$$
\begin{bmatrix}
-1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

---

### 🔧 **Aplikasi Refleksi Matriks**:

* Grafika komputer (cermin objek)
* Geometri transformasi
* Simulasi fisika (pantulan)




```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik asal (domain)
koordinats = np.array([[1, 0, -1], [1, 0, -1]])
x = koordinats[0, :]
y = koordinats[1, :]

# Matriks refleksi terhadap sumbu X
B = np.array([[1, 0],
              [0, -1]])
B_trans = B @ koordinats
x_LT2 = B_trans[0, :]
y_LT2 = B_trans[1, :]

# Membuat gambar dan sumbu-sumbu
fig, ax = plt.subplots()

# Plot titik-titik
ax.plot(x, y, 'ro', label='Asal (domain)')
ax.plot(x_LT2, y_LT2, 'bo', label='Bayangan (codomain)')

# Menghubungkan titik-
ax.plot(x, y, 'r--')
ax.plot(x_LT2, y_LT2, 'b')

# Gambar sumbu-sumbu
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")

# Properti plot
ax.grid(True)
ax.axis([-2, 2, -2, 2])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap sumbu X")
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik asal (domain)
koordinats = np.array([[1, 0, -1], [1, 0, -1]])
x = koordinats[0, :]
y = koordinats[1, :]

# Matriks refleksi terhadap garis y = x
B = np.array([[0, 1],
              [1, 0]])
B_trans = B @ koordinats
x_LT2 = B_trans[0, :]
y_LT2 = B_trans[1, :]

# Membuat gambar dan sumbu-sumbu
fig, ax = plt.subplots()

# Plot titik-titik
ax.plot(x, y, 'ro', label='Asal (domain)')
ax.plot(x_LT2, y_LT2, 'bo', label='Bayangan (codomain)')

# Menghubungkan titik-titik asal dan bayangan
for i in range(x.shape[0]):
    ax.plot([x[i], x_LT2[i]], [y[i], y_LT2[i]], 'gray', ls=':')

# Gambar garis y = x
t = np.linspace(-2, 2, 100)
ax.plot(t, t, 'g--', label='Garis y = x')

# Gambar sumbu-sumbu
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")

# Properti plot
ax.grid(True)
ax.axis([-2, 2, -2, 2])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap garis y = x")
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik asal (domain)
koordinats = np.array([[1, 0, -1], [3, 2, 1]])
x = koordinats[0, :]
y = koordinats[1, :]

# Langkah 1: Translasi ke bawah sejauh 2 (y -> y - 2)
T_down = np.array([[1, 0, 0],
                   [0, 1, -2],
                   [0, 0, 1]])

# Langkah 2: Refleksi terhadap sumbu X
Reflect_X = np.array([[1, 0, 0],
                      [0, -1, 0],
                      [0, 0, 1]])

# Langkah 3: Translasi ke atas sejauh 2 (y -> y + 2)
T_up = np.array([[1, 0, 0],
                 [0, 1, 2],
                 [0, 0, 1]])

# Gabungkan semua transformasi (T_up @ Reflect_X @ T_down)
Transform = T_up @ Reflect_X @ T_down

# Ubah koordinat ke bentuk homogen (3xN)
homogen = np.vstack((koordinats, np.ones((1, koordinats.shape[1]))))

# Lakukan transformasi
transformed = Transform @ homogen
x_LT = transformed[0, :]
y_LT = transformed[1, :]

# Visualisasi
fig, ax = plt.subplots()
ax.plot(x, y, 'ro', label='Asal (domain)')
ax.plot(x_LT, y_LT, 'bo', label='Bayangan (codomain)')

# Hubungkan tiap titik asal ke bayangannya
for i in range(x.shape[0]):
    ax.plot([x[i], x_LT[i]], [y[i], y_LT[i]], 'gray', ls=':')

# Garis refleksi y = 2
ax.axhline(y=2, color='g', ls='--', label='y = 2')

# Sumbu-sumbu
ax.axvline(x=0, color='k', ls=':')
ax.axhline(y=0, color='k', ls=':')

# Properti plot
ax.grid(True)
ax.set_aspect('equal')
ax.axis([-2, 2, -1, 5])
ax.set_title("Refleksi terhadap garis y = 2")
ax.legend()

plt.show()
```
