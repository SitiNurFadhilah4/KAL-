# Sistem Persamaan Linier

## pengertian sistem persamaan linier
Sistem Persamaan Linier (SPL) adalah kumpulan dari dua atau lebih persamaan linier yang memiliki variabel yang sama dan harus diselesaikan secara bersamaan.

contoh persamaan 3 variabel:

$x + y + z = 3$
$2x - y + z = 4$
$x + 2y - z = 1$

bentuk umum SPL:

$a‚ÇÅx + b‚ÇÅy + c‚ÇÅz = d$‚ÇÅ
$a‚ÇÇx + b‚ÇÇy + c‚ÇÇz = d‚ÇÇ$
$a‚ÇÉx + b‚ÇÉy + c‚ÇÉz = d‚ÇÉ$

## Jenis Penyelesaian SPL ada  3:
1. Tunggal: Hanya satu solusi (potong di satu titik)
2. Tak Hingga: Banyak solusi (grafik berupa garis/himpunan yang berimpit)
3. Tidak Ada Solusi: Tidak ada titik temu (garis sejajar/tidak konsisten)

## metode penyelesaian SPL
a. metode substitusi

cara penyelesaaian menggunakan metode substitusi:
Pilih salah satu persamaan, kemudian nyatakan salah satu variabel dalam bentuk variabel lainnya.
Substitusikan hasil tersebut ke persamaan lain.
Selesaikan persamaan untuk menemukan nilai variabel yang tersisa.

b. metode eliminasi

cara penyelesaaian menggunakan metode eliminasi:
Samakan salah satu koefisien variabel dengan mengalikan persamaan jika diperlukan.
Kurangkan atau jumlahkan kedua persamaan untuk menghilangkan salah satu variabel.
Selesaikan persamaan yang tersisa.

c.eliminasi gauss

Eliminasi Gauss bertujuan mengubah matriks augmented menjadi bentuk eselon baris atas (upper triangular form) menggunakan operasi baris elementer.
cara penyelesaaian menggunakan metode eliminasi:
Bentuk segitiga atas (eliminasi ke bawah),
Ubah elemen-elemen di bawah diagonal utama menjadi nol menggunakan operasi baris.
Substitusi mundur (back-substitution),
Selesaikan nilai variabel dari persamaan yang sudah disederhanakan.




## penjelasan Tugas 

### jika berhimpitam mempunyai banyak solusi

persamaan:
$x+2y+3z=6
2x+4y+6z=12
3x+6y+9z=18.$    

Langkah 1: Analisis Koefisien
Perhatikan bahwa setiap persamaan merupakan kelipatan dari persamaan pertama:

Baris kedua adalah 2 kali baris pertama.
Baris ketiga adalah 3 kali baris pertama.
Artinya, ketiga persamaan tersebut sebenarnya merepresentasikan persamaan yang sama. Oleh karena itu, sistem persamaan ini tidak memiliki solusi unik tetapi memiliki banyak solusi (tak hingga solusi).

Langkah 2: Representasi dalam Matriks
Sistem persamaan di atas dapat ditulis dalam bentuk matriks augmented:

$\begin{bmatrix} 
1 & 2 & 3 & \vert & 6 \\ 
2 & 4 & 6 & \vert & 12 \\ 
3 & 6 & 9 & \vert & 18 
\end{bmatrix}$

Kita lakukan eliminasi baris:

Baris kedua dikurangkan 2 kali baris pertama:

$R_2 \leftarrow R_2 - 2R_1$

Hasilnya:

$\begin{bmatrix} 
1 & 0 & 3 & \vert & 6 \\ 
2 & 0 & 6 & \vert & 0 \\ 
3 & 0 & 9 & \vert & 18 
\end{bmatrix}$

Baris ketiga dikurangkan 3 kali baris pertama:

$R_3 \leftarrow R_3 - 3R_1$

Hasilnya:

$\begin{bmatrix} 
1 & 0 & 0 & \vert & 6 \\ 
2 & 0 & 0 & \vert & 0 \\ 
3 & 0 & 0 & \vert & 0 
\end{bmatrix}$
Langkah 3: Menentukan Solusi
Dari matriks terakhir, kita hanya memiliki satu persamaan independen:

$x+2y+3z=6$
Karena ada tiga variabel tetapi hanya satu persamaan, maka kita bisa menyatakan solusi dalam bentuk parametrik:

Biarkan y = t dan z =s sebagai parameter bebas.
Dari persamaan pertama:
$x=6‚àí2y‚àí3z$

Substitusi y=t dan z=s:
$x=6‚àí2t‚àí3s$


Kesimpulan
Solusi dari sistem persamaan ini dalam bentuk parametrik adalah:

$\begin{aligned}
x &= 6 - 2t - 3s \\
y &= t \\
z &= s
\end{aligned}$

dengan t,s sebagai bilangan real bebas.
Karena ada dua parameter bebas (t dan s),ini menunjukkan bahwa sistem memiliki tak hingga banyak solusi yang membentuk bidang dalam ruang tiga dimensi.

### jika tidak berpotongan tidak ada solusi

persamaan:
$x+y+z=1
x+y+z=3
x+y+z=5$

Langkah 1: Perhatikan Struktur Persamaan
Semua persamaan memiliki bentuk yang sama, yaitu $x+y+z$, tetapi dengan hasil yang berbeda (1, 3, dan 5). Ini tidak mungkin terjadi karena satu ekspresi tidak dapat memiliki tiga nilai berbeda pada saat yang sama.

Langkah 2: Kurangkan Persamaan
Misalnya, kita kurangkan persamaan kedua dengan persamaan pertama:
$\begin{align*}
(x + y + z) - (x + y + z) &= 3 - 1 \\
0 &= 2
\end{align*}$

Ini merupakan kontradiksi (pernyataan yang salah).

Lakukan hal yang sama untuk persamaan ketiga dengan persamaan pertama:
$\begin{align*}
(x + y + z) - (x + y + z) &= 5 - 1 \\
0 &= 4
\end{align*}$

Ini juga kontradiksi.

Kesimpulan
Sistem ini adalah inkonsisten, artinya tidak ada solusi.
Dalam konteks geometri, ini menunjukkan bahwa tiga bidang yang direpresentasikan oleh persamaan tersebut sejajar tetapi tidak berpotongan, sehingga tidak ada titik perpotongan yang memenuhi semua persamaan sekaligus.

Langkah 2: Substitusi dan Eliminasi Lanjutan
Kita punya sistem:


$4x + 3y = 8$

$-3y + z = 2$

$z = 3y + 2$

$3x - y = 4 \quad \text{(Persamaan D)}$

Langkah 3: Menyelesaikan Sistem Dua Variabel

Kita punya:

$x + 4y = 4$

$3x - y = 4$

$3(x + 4y) = 3(4)$

$3x + 12y = 12$

$13y = 8$


$y = \frac{8}{13}$

$x + 4 \left(\frac{8}{13}\right) = 4$

$x = 4 - \frac{32}{13}$

$x = \frac{52}{13} - \frac{32}{13}$

$x = \frac{20}{13}$

$z = 3\left(\frac{8}{13}\right) + 2$

$(x, y, z) = \left( \frac{20}{13}, \frac{8}{13}, \frac{50}{13} \right)$

## Transformasi Matriks 

Transformasi matriks adalah fungsi yang mengubah vektor dari satu ruang vektor ke ruang vektor lain menggunakan perkalian matriks.

Misalnya, jika ùê¥ adalah matriks ùëö √ó ùëõ dan $\vec{x} ‚àà ùëÖ^ùëõ$ maka transformasi ùëá didefinisikan sebagai:

$T(\vec{x}) = A\vec{x}$

Outputnya adalah vektor $\vec{y}‚àà ùëÖ^ùëö.$
Transformasi seperti ini disebut transformasi linear, dan merupakan bagian penting dalam pemetaan antara ruang vektor.

### a.  Perkalian Matriks-Vektor
Fokus pada vektor 2D $(ùëÖ^2)$ untuk mempermudah visualisasi.
Matriks 2 √ó 2 bisa digunakan untuk mengubah vektor 2D dan tetap menghasilkan vektor 2D.

Contoh matriks:
$A = \begin{bmatrix} 1 & 4 \\ 2 & 3 \end{bmatrix}$

jika $\vec{x}$ = $\begin{bmatrix} 1 \\ 0 \end{bmatrix}$ maka:

$A\vec{x} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$

Visualisasi digunakan untuk menunjukkan bagaimana vektor sebelum dan sesudah transformasi berubah pada bidang kartesius.

### b. Transformasi Dasar
Beberapa contoh transformasi umum yang bisa direpresentasikan oleh matriks:

Refleksi terhadap sumbu-x: $\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$

Refleksi terhadap sumbu-y: $\begin{bmatrix} -1 & 0 \\ 0 & 1 \end{bmatrix}$

Rotasi berlawanan arah jarum jam sebesar ùúÉ : $\begin{bmatrix} \cos \theta & -\sin \theta \\ \sin \theta & \cos \theta \end{bmatrix}$

Peregangan horizontal atau vertikal: 

Horizontal: $\begin{bmatrix}
k & 0 \\
0 & 1
\end{bmatrix}$ , 
Vertikal:¬†$\begin{bmatrix}
1 & 0 \\
0 & k
\end{bmatrix}$

### c. Komposisi Transformasi
Transformasi dapat dikombinasikan dengan mengalikan matriks secara berurutan.
Misal, transformasi T1 diikuti T2:
$A = A_2 \cdot A_1$

Maka transformasi total $T(\vec{x}) = A_2 (A_1 \vec{x})$

# tugas 1 membuktikan $T(v_1, v_2) = (v_1 + v_2, v_1)$ adalah transformasi linier 

Untuk membuktikan bahwa transformasi $T(v_1, v_2) = (v_1 + v_2, v_1)$ adalah transformasi linier, kita harus menunjukkan bahwa:

1. Penjumlahan vektor dipertahankan: $T(u+v)=T(u)+T(v)$ 
2. Perkalian skalar dipertahankan: $T(c‚ãÖv)=c‚ãÖT(v)$

Misalkan: $u = (u_1, u_2),\quad v = (v_1, v_2),\quad c \in \mathbb{R}$

1.  Cek Penjumlahan: 

$T(u + v) = T((u_1 + v_1, u_2 + v_2)) = ((u_1 + v_1) + (u_2 + v_2),\ u_1 + v_1)$

$= (u_1 + u_2 + v_1 + v_2,\ u_1 + v_1)$

Sementara: 

$T(u) = (u_1 + u_2,\ u_1),\quad T(v) = (v_1 + v_2,\ v_1)$

$T(u) + T(v) = (u_1 + u_2 + v_1 + v_2,\ u_1 + v_1)$

Sama, jadi penjumlahan dipertahankan.

2. Cek Perkalian Skalar:

$T(c \cdot v) = T((cv_1, cv_2)) = (cv_1 + cv_2,\ cv_1) = c(v_1 + v_2,\ v_1)$

$= c \cdot T(v)$

Sama, jadi perkalian skalar juga dipertahankan.

Karena kedua sifat dipenuhi, maka:

$T(v_1, v_2) = (v_1 + v_2,\ v_1)$
adalah transformasi linier.

# Refleksi

**Refleksi matriks** (reflection matrix) adalah **matriks transformasi** yang mencerminkan (memantulkan) titik-titik dalam ruang terhadap **sumbu, garis, bidang, atau titik tertentu**.

Refleksi mengubah arah satu atau lebih komponen vektor, seperti bayangan di cermin.

---

### ‚úÖ **Repleksi pada 2D (dua dimensi)**:

1. **Refleksi terhadap sumbu-x**:

$$
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
x \\
-y
\end{bmatrix}
$$

‚Üí Membalik titik terhadap sumbu-x (nilai y dibalik, x tetap).

2. **Refleksi terhadap sumbu-y**:

$$
\begin{bmatrix}
-1 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
-x \\
y
\end{bmatrix}
$$

‚Üí Membalik titik terhadap sumbu-y.

3. **Refleksi terhadap garis y = x**:

$$
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
y \\
x
\end{bmatrix}
$$

‚Üí Menukar posisi x dan y.

---

### ‚úÖ ** Repleksi pada 3D (tiga dimensi)**:

Refleksi bisa dilakukan terhadap bidang seperti $x=0$, $y=0$, atau $z=0$. Contoh refleksi terhadap bidang $x=0$:

$$
\begin{bmatrix}
-1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

---

### üîß **Aplikasi Refleksi Matriks**:

* Grafika komputer (cermin objek)
* Geometri transformasi
* Simulasi fisika (pantulan)




```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik asal (domain)
koordinats = np.array([[1, 0, -1], [1, 0, -1]])
x = koordinats[0, :]
y = koordinats[1, :]

# Matriks refleksi terhadap sumbu X
B = np.array([[1, 0],
              [0, -1]])
B_trans = B @ koordinats
x_LT2 = B_trans[0, :]
y_LT2 = B_trans[1, :]

# Membuat gambar dan sumbu-sumbu
fig, ax = plt.subplots()

# Plot titik-titik
ax.plot(x, y, 'ro', label='Asal (domain)')
ax.plot(x_LT2, y_LT2, 'bo', label='Bayangan (codomain)')

# Menghubungkan titik-
ax.plot(x, y, 'r--')
ax.plot(x_LT2, y_LT2, 'b')

# Gambar sumbu-sumbu
ax.axvline(x=0, color="k", ls=":")
ax.axhline(y=0, color="k", ls=":")

# Properti plot
ax.grid(True)
ax.axis([-2, 2, -2, 2])
ax.set_aspect('equal')
ax.set_title("Refleksi terhadap sumbu X")
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Titik-titik asal (domain)
koordinats = np.array([[1, 0, -1], [3, 2, 1]])
x = koordinats[0, :]
y = koordinats[1, :]

# Langkah 1: Translasi ke bawah sejauh 2 (y -> y - 2)
T_down = np.array([[1, 0, 0],
                   [0, 1, -2],
                   [0, 0, 1]])

# Langkah 2: Refleksi terhadap sumbu X
Reflect_X = np.array([[1, 0, 0],
                      [0, -1, 0],
                      [0, 0, 1]])

# Langkah 3: Translasi ke atas sejauh 2 (y -> y + 2)
T_up = np.array([[1, 0, 0],
                 [0, 1, 2],
                 [0, 0, 1]])

# Gabungkan semua transformasi (T_up @ Reflect_X @ T_down)
Transform = T_up @ Reflect_X @ T_down

# Ubah koordinat ke bentuk homogen (3xN)
homogen = np.vstack((koordinats, np.ones((1, koordinats.shape[1]))))

# Lakukan transformasi
transformed = Transform @ homogen
x_LT = transformed[0, :]
y_LT = transformed[1, :]

# Visualisasi
fig, ax = plt.subplots()
ax.plot(x, y, 'ro', label='Asal (domain)')
ax.plot(x_LT, y_LT, 'bo', label='Bayangan (codomain)')

# Hubungkan tiap titik asal ke bayangannya
for i in range(x.shape[0]):
    ax.plot([x[i], x_LT[i]], [y[i], y_LT[i]], 'gray', ls=':')

# Garis refleksi y = 2
ax.axhline(y=2, color='g', ls='--', label='y = 2')

# Sumbu-sumbu
ax.axvline(x=0, color='k', ls=':')
ax.axhline(y=0, color='k', ls=':')

# Properti plot
ax.grid(True)
ax.set_aspect('equal')
ax.axis([-2, 2, -1, 5])
ax.set_title("Refleksi terhadap garis y = 2")
ax.legend()

plt.show()
```

Eigenvector dari sebuah matriks adalah vektor non-nol yang arah atau orientasinya tidak berubah ketika dilakukan transformasi linear oleh matriks tersebut. Yang berubah hanyalah skalanya.

#### Eigenvalue (Nilai Eigen)

Eigenvalue adalah skalar yang menunjukkan berapa besar skala perubahan (perbesaran atau pengecilan) yang dialami oleh eigenvector setelah transformasi. 

Jika diberikan sebuah matriks persegi A dan vektor v, maka:

$A\mathbf{v} = \lambda \mathbf{v}$

Keterangan:

$A$ = matriks persegi (n √ó n)

$v$ = eigenvector (vektor kolom, v ‚â† 0)

$\lambda$ = eigenvalue

#### definisi Orthonormal dan ortogonal

##### Orthonormal
Suatu himpunan dikatakan ortonormal jika ortogonal dan setiap vektor adalah vektor satuan (norma sama dengan 1).
Artinya, vektor-vektor tersebut saling tegak lurus dan memiliki panjang (norma) satu.
Setiap himpunan ortogonal dapat dikonversi menjadi himpunan ortonormal dengan menormalisasikan setiap vektornya. 

Secara matematis:

$\mathbf{u}_i \cdot \mathbf{u}_j = 0 \quad \text{jika} \quad i \neq j \quad \text{(ortogonal)}$

$\| \mathbf{u}_i \| = 1 \quad \text{untuk semua } i \quad \text{(norma = 1)}$

##### ortogonal

Dua vektor dikatakan ortogonal jika sudut antara keduanya adalah 90 derajat atau œÄ/2 radian. Secara matematis, dua vektor $\vec{u} \text{ dan } \vec{v}$ adalah ortogonal jika:

$\vec{u} \cdot \vec{v} = 0$

Artinya, hasil dari dot product (perkalian titik) mereka adalah nol.


#### Mencari Ortonormal dan ortogonal

##### Mencari ortogonal

Sebuah matriks A adalah ortogonal jika:

$A^T \cdot A = I$

Dimana $A^T$ adalah transpose dari A dan I adalah matriks identitas.

Coba kita hitung:

$A^T = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}$

$A^T \cdot A =$
$\begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
\cdot$
$\begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
=$
$\begin{bmatrix} 5 & 4 \\ 4 & 5 \end{bmatrix}$

Karena hasilnya bukan matriks identitas, maka A bukan matriks ortogonal.

##### Mencari ortonormal

Syarat ortonormal:

Vektor-vektor kolomnya saling ortogonal (dot product = 0)

Panjang (norma) masing-masing vektor kolom = 1

Kolom-kolom dari $A:$

$v‚ÇÅ = [2, 1]$

$v‚ÇÇ = [1, 2]$

Dot product: $2‚ãÖ1+1‚ãÖ2=4 ‚Üí$ tidak ortogonal

Norm $v_1 : \quad \sqrt{2^2 + 1^2} = \sqrt{5} \neq 1$

Jadi, kolom-kolom A juga tidak ortonormal.

##### menjadikan vektor yang didapat menjadi ortonormal(panjang vektor menjadi satu satuan)

Matriks $A$ :
$A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}$

Kolom-vektor:
$\vec{v}_1 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}$

$\vec{v}_2 = \begin{bmatrix} 1 \\ 2 \end{bmatrix}$

Langkah 1 dengan cara Normalisasi :

Normalisasi dilakukan dengan: 
$\hat{v} = \frac{\vec{v}}{\|\vec{v}\|}$

dimana $‚à•v‚à•$ adalah norma (panjang) vektor.

Normalisasi $\vec{v}_1$ 

$\|v_1\| = \sqrt{2^2 + 1^2} = \sqrt{5}$

$\hat{v}_1 = \frac{1}{\sqrt{5}} \begin{bmatrix} 2 \\ 1 \end{bmatrix} = \begin{bmatrix} \frac{2}{\sqrt{5}} \\[6pt] \frac{1}{\sqrt{5}} \end{bmatrix}$

Normalisasi $\vec{v}_2$ 

$\|v_2\| = \sqrt{1^2 + 2^2} = \sqrt{5}$

$\hat{v}_2 = \frac{1}{\sqrt{5}} \begin{bmatrix} 1 \\ 2 \end{bmatrix} = \begin{bmatrix} \frac{1}{\sqrt{5}} \\[6pt] \frac{2}{\sqrt{5}} \end{bmatrix}$

$Hasil:$

Dua vektor ortonormal hasil normalisasi:

$\hat{v}_1 = \begin{bmatrix} \frac{2}{\sqrt{5}} \\[6pt] \frac{1}{\sqrt{5}} \end{bmatrix}$
$, \hat{v}_2 = \begin{bmatrix} \frac{1}{\sqrt{5}} \\[6pt] \frac{2}{\sqrt{5}} \end{bmatrix}$

$Keterangan :$

Panjang masing-masing vektor = 1 (sudah normal).

Namun dot product keduanya $= \frac{2}{\sqrt{5}} \cdot \frac{1}{\sqrt{5}} + \frac{1}{\sqrt{5}} \cdot \frac{2}{\sqrt{5}} = \frac{4}{5} \neq 0 ,$  jadi belum saling ortogonal.

```python
import numpy as np

A = np.array([[2, 1],
              [1, 2]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("Matriks A:\n", A)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors (kolom-kolomnya adalah eigenvector):\n", eigenvectors)


for i in range(len(eigenvalues)):
    Œª = eigenvalues[i]
    v = eigenvectors[:, i]
    
    Av = A @ v
    Œªv = Œª * v
    
    print(f"\nVerifikasi untuk eigenvalue Œª = {Œª:.2f}:")
    print("A @ v =", Av)
    print("Œª * v =", Œªv)
    print("Sama?", np.allclose(Av, Œªv))
```

Untuk menghitung perkalian silang (cross product) antara dua vektor 3 dimensi:

$\vec{u} = (1,1,1),$   $\vec{v} =(2,1,1)$

Langkah-langkah:

Gunakan rumus determinan matriks sebagai berikut:

$\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
1 & 1 & 1 \\
2 & 1 & 1
\end{vmatrix}$

Hitung komponennya:

$ Komponen  {i}:$

$\\{i} \cdot (1 \cdot 1 - 1 \cdot 1) = \hat{i} \cdot (1 - 1) = 0\hat{i}$

$Komponen j:$

$\\{j} \cdot (1 \cdot 1 - 1 \cdot 2) = -\hat{j} \cdot (1 - 2) = -\hat{j} \cdot (-1) = 1\hat{j}$

$Komponen k:$

$k‚ãÖ(1‚ãÖ1‚àí1‚ãÖ2)=k‚ãÖ(1‚àí2)=‚àí1k$

$Hasil akhir:$

$\vec{u} \times \vec{v} = 0\hat{i} + 1\hat{j} - 1\hat{k} = (0, 1, -1)$

##### Soal 1

1. Tentukan luas jajaran genjang yang ditentukan oleh:

$\vec{u} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 2 \\ 1 \end{bmatrix}$

Langkah-langkah:

1). Bentuk matriks dari vektor $\vec{u}$ dan $\vec{v}$

2). Hitung determinan dari matriks:

$\begin{vmatrix}
1 & 2 \\
2 & 1
\end{vmatrix}
= (1 \cdot 1) - (2 \cdot 2) = 1 - 4 = -3$

3). Ambil nilai absolut dari hasil determinan:

$|{-3}| = 3$

Jadi, luas jajaran genjang adalah 3 satuan luas.

##### Soal 2

2. Tentukan luas jajaran genjang yang ditentukan oleh:

$\vec{u} = \begin{bmatrix} 2 \\ 0 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 0 \\ 3 \end{bmatrix}$

Langkah-langkah:

1). Bentuk matriks dari kedua vektor.
2). Hitung determinan:

$\begin{vmatrix}
2 & 0 \\
0 & 3
\end{vmatrix}
= (2 \cdot 3) - (0 \cdot 0) = 6$

3).Ambil nilai absolut (meskipun sudah positif):

$‚à£6‚à£= 6$

Jadi, luas jajaran genjang adalah 6 satuan luas.

##### Soal 3

3. Tentukan luas segitiga dengan titik-titik:

$(0,0,0),(1,3,‚àí1),(2,1,1)$

Langkah-langkah:

1). Tentukan vektor dari titik A ke B dan A ke C:

$\vec{AB} = \begin{pmatrix} 1 \\ 3 \\ -1 \end{pmatrix}, \quad
\vec{AC} = \begin{pmatrix} 2 \\ 1 \\ 1 \end{pmatrix}$

2). Hitung hasil perkalian silang $\vec{AB} \times \vec{AC}:$

$\vec{AB} \times \vec{AC} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
1 & 3 & -1 \\
2 & 1 & 1
\end{vmatrix}
= 4\hat{i} - 3\hat{j} - 5\hat{k}$

4). Hitung besar vektor hasil cross product:

$\left| \vec{AB} \times \vec{AC} \right| =
\sqrt{4^2 + (-3)^2 + (-5)^2}
= \sqrt{16 + 9 + 25}
= \sqrt{50}$

5). Hitung luas segitiga:

$L = \frac{1}{2} \cdot \left| \vec{AB} \times \vec{AC} \right| = \frac{1}{2} \cdot \sqrt{50} = \frac{\sqrt{50}}{2} \approx 3.54$
